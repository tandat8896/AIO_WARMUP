```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("advertising.csv")
df.head(5)
```

```python
X=df[["TV","Radio","Newspaper"]].values
y=df[["Sales"]].values.reshape(-1,1)
```

```python
X = (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
def train_linear_regression(X,y,lr=0.00001 , epochs = 100,tolerance=1e-6):
    n_samples, n_features = X.shape
    loss_history = []
    w = np.zeros((n_features,1))
    for i in range(epochs):
        y_pred = X @ w
        error = y_pred - y 
        loss = np.mean(error**2)
        loss_history.append(loss)

        grad = 2 * X.T @ error/n_samples

        w = w - lr * grad
        if i > 0 and abs(loss_history[i] - loss_history[i-1]) < tolerance:
            print(f"Đã dừng sớm tại epoch {i} do sự thay đổi loss rất nhỏ")
            break
    return w , loss_history


    
```

```python
w, losses = train_linear_regression(X, y,lr = 0.00001 , epochs = 1000)
print("Loss sau mỗi epoch:")
for i, l in enumerate(losses):
    print(f"Epoch {i+1}: {l}")
```

```python
import streamlit as st 
import matplotlib.pyplot as plt 
```

```python
st.title("Huấn Luyện Linear_Regression với numpy ")
lr = st.slider("learning rate",0.001,0.1,0.01)
epochs = st.slider("so epochs",10,1000,200)

if st.button("Train model"):
    w,b,loss_history= train_linear_regression(X,y,lr,epochs , tolerance)
    st.subheader("Loss theo epoch")

    fig, ax = plt.subplots()
    ax.plot(loss_history)
    ax.set_xlabel("Epoch")
    ax.set_ylabel("MSE Loss")
    st.pyplot(fig)

    st.write("Trọng số cuối cùng (w):", w.flatten())
    st.write("Bias (b):", b)
```
