```python
lst_data = []
for i in range(1, 13):
    if i % 2 == 0:
        lst_data.append(i)
print(lst_data)
```

```python
for i in range(len(lst_data) - 1, -1, -1):
    if lst_data[i] % 3 == 0:
        lst_data.pop(i)

print(lst_data)
```

```python
lst_data.extend(range(1, 4))

print(lst_data)
```

```python
index = 3
for i in range(8, 5, -1):
    lst_data.insert(index, i)

print(lst_data)
```

```python
for i in range(len(lst_data)):
    if lst_data[i] % 2 == 0 or lst_data[i] % 5 == 0:
        lst_data[i] = 0
print(lst_data)
```
