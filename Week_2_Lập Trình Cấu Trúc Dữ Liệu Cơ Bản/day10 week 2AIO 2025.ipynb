```python
import math
def create_position_matrix(seq_length,embed_size):
  #khởi tạo postion matrix bằng 0 qua các cột của embed_size và qua từng từ trong vocab
  position_matrix=[[0 for i in range(embed_size)] for j in range(seq_length)]
  #chạy vòng lặp cho từng từ
  for pos in range(seq_length):
    #trong từng từ lấy các cột là các chiều của vector embed_size
    for i in range(embed_size):
      #nếu vị trí embed_size chẳn thì dùng
      if i % 2 == 0:
        position_matrix[pos][i] = math.sin(pos / (10000 ** ((2 * i) / embed_size)))
      else :
      #nếu vị trí em_size
        position_matrix[pos][i] = math.cos(pos / (10000 ** ((2 * i) / embed_size)))
  return position_matrix
```

```python
seq_length =10
embed_size = 16
position_matrix = create_position_matrix(seq_length,embed_size)
print(position_matrix)
for i in position_matrix:
  print(i)
```
