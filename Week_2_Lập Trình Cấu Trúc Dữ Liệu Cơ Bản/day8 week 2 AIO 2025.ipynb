```python
# Bài6
mat_a = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

mat_b = [
    [2, 4, 6],
    [1, 3, 5],
    [1, 0, 1]
]

#  tổng A + B
sum_matrix = []
for i in range(3):  #1 hàng trong ma trận
    row = []
    for j in range(3):  #1 phần tử trong 1 hàng
        row.append(mat_a[i][j] + mat_b[i][j])
    sum_matrix.append(row)

#hiệu A - B
diff_matrix = []
for i in range(3):  #1 hàng trong ma trận
    row = []
    for j in range(3):  #1 phần tử trong 1 hàng
        row.append(mat_a[i][j] - mat_b[i][j])
    diff_matrix.append(row)

# Tính dot product A * B
dot_product = []
for i in range(3):#1 hàng trong ma trận
    row = []
    for j in range(3):
        sum_value = 0
        for k in range(3):
            sum_value += mat_a[i][k] * mat_b[k][j]
        row.append(sum_value)
    dot_product.append(row)

print(sum_matrix)
print(diff_matrix)
print(dot_product)
```

```python
#bài 8
def normalized(lst_data):
  min_value = min(lst_data)
  max_value = max(lst_data)
  return [(x-min_value)/(max_value-min_value) for x in lst_data]

test_cases = [
    [11, 18, 24, 30, 36],
    [50, 100, 150, 200, 250],
    [3, 5, 7, 9, 11]
]

for i,test in enumerate(test_cases):
  result = normalized(test)
  print(result)
```

```python
#Bài 10
def dot_product(v1, v2):
    result = 0
    for i in range(len(v1)):
        result += v1[i] * v2[i]
    return result

# Test cases
test_cases = [
    ([1, 2, 3], [4, 5, 6]),
    ([2, 4, 6], [1, 3, 5]),
    ([0, 1, 2], [3, 4, 5])
]


for i, (v1, v2) in enumerate(test_cases):
    result = dot_product(v1, v2)
    print(result)
```
