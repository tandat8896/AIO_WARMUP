```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def derivative_chainrule(f,x,h=1e-5):
  numerical = (f(x+h)-f(x))
  derivative = numerical/h
  return derivative

# Hàm tính đạo hàm số học
def calculate_derivative(f, x, h=1e-5):
    return (f(x + h) - f(x)) / h

# Hàm vẽ đồ thị Chain Rule
def plot_chain_rule(f, g, x_range=(-2, 2), title="Đạo hàm Chain Rule"):
    x_vals = np.linspace(x_range[0], x_range[1], 400)

    # Tính giá trị của g(x) và f(g(x))
    g_vals = g(x_vals)
    f_vals = f(g_vals)

    # Tính đạo hàm theo Chain Rule: f'(g(x)) * g'(x)
    f_prime_vals = []
    for g_x in g_vals:
      f_prime_vals.append(calculate_derivative(f, g_x))  # Đạo hàm của f tại g(x)

    g_prime_vals = []
    for x in x_vals:
      g_prime_vals.append(calculate_derivative(g, x))  # Đạo hàm của g tại x

    chain_rule_derivative_vals = []
    for f_p, g_p in zip(f_prime_vals, g_prime_vals):
      chain_rule_derivative_vals.append(f_p * g_p)  # Áp dụng quy tắc Chain Rule

    #f_prime_vals = [calculate_derivative(f, g_x) for g_x in g_vals]
    #g_prime_vals = [calculate_derivative(g, x) for x in x_vals]
    #chain_rule_derivative_vals = [f_p * g_p for f_p, g_p in zip(f_prime_vals, g_prime_vals)]

    # Vẽ đồ thị của f(g(x)) và đạo hàm
    plt.figure(figsize=(8, 5))
    plt.plot(x_vals, f_vals, label="f(g(x))", color='blue')
    plt.plot(x_vals, chain_rule_derivative_vals, label="f'(x) theo Chain Rule", color='red', linestyle="--")

    plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
    plt.axvline(0, color='black', linewidth=0.5, linestyle='--')

    plt.xlabel("x")
    plt.ylabel("Giá trị")
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()


```

```python
functions = [
    (lambda x: np.exp(x), lambda x: x**3 - 2*x + 1, (-2, 2), "Bài 1: Đạo hàm của hàm mũ hợp"),
    (lambda x: np.log(x), lambda x: 5*x**2 + 3*x + 1, (0.1, 3), "Bài 2: Đạo hàm của hàm logarit hợp"),
    (lambda x: np.sqrt(x), lambda x: x**4 + 2*x**2 + 1, (-2, 2), "Bài 3: Đạo hàm của hàm căn bậc hai hợp"),
    (lambda x: np.sin(x), lambda x: x**2 - 3*x + 2, (-2, 2), "Bài 4: Đạo hàm của hàm lượng giác hợp"),
    (lambda x: np.exp(x), lambda x: np.log(x**2 + 4), (0.1, 3), "Bài 5: Đạo hàm của hàm hợp có nhiều lớp")
]

# Chạy và vẽ đồ thị cho từng bài toán
for f, g, x_range, title in functions:
    plot_chain_rule(f, g, x_range, title)
```
