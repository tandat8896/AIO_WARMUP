```python

# Hàm lấy một cột từ dữ liệu
def get_colum(data, index):
    N = len(data)
    result = []
    for i in range(N):
        result.append(data[i][index])
    return result

# Hàm đọc và chuẩn bị dữ liệu
def prepare_data(file_name):
    data = np.genfromtxt(file_name, delimiter=',', skip_header=1)
    tv_data = get_colum(data, 0)
    radio_data = get_colum(data, 1)
    newspaper_data = get_colum(data, 2)
    sales_data = get_colum(data, 3)
    X = [tv_data, radio_data, newspaper_data]
    y = sales_data
    return X, y  # Chỉ trả về X và y


# Đọc file dữ liệu
file_name = "/content/advertising.csv"
data = prepare_data(file_name)



```

```python
#cau4
def compute_gradient_wi(xi, y, y_hat):
  dl_dwi = -2 * (y_hat - y) * xi
  return dl_dwi
```

```python
#cau4
def compute_gradient_b( y, y_hat):
  dl_db = 2 * (y_hat - y)
  return dl_db
```
