```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def calculate_derivative(f,x,h=1e-5):
  numerical = (f(x+h)-f(x))
  derivative = numerical/h
  return derivative
```

```python
def plot_function_and_derivative(f, x_range=(-2,2) , title="Đồ thị f(x)và f'(x)"):
  x_vals = np.linspace(x_range[0],x_range[1],400)
  y_vals = f(x_vals)
  y_derivative_vals = np.array([calculate_derivative(f, xi) for xi in x_vals])
  plt.figure(figsize=(8,5))
  plt.plot(x_vals,y_vals,label="f(x)",color='blue')
  plt.plot(x_vals,y_derivative_vals,label="f'(x) (numerical)", color='red', linestyle="--")


  plt.axhline(0 , color='black', linewidth = 0.5 , linestyle='--')
  plt.axvline(0 , color='black', linewidth = 0.5 , linestyle='--')
  plt.grid(True)
  plt.xlabel("x")
  plt.ylabel("f(x) & f'(x)")
  plt.title(title)
  plt.grid()
  plt.legend()
  plt.show()
```

```python
functions = [
    (lambda x: x**3 - 3*x**2 + 4*x - 2, (-2, 3), "Bài 2: Đạo hàm của hàm bậc ba"),
    (np.exp, (-2, 2), "Bài 3: Đạo hàm của hàm số mũ"),
    (np.log, (0.1, 3), "Bài 4: Đạo hàm của hàm logarit"),
    (np.sin, (-np.pi, np.pi), "Bài 5: Đạo hàm của hàm sin(x)"),
    (np.cos, (-np.pi, np.pi), "Bài 6: Đạo hàm của hàm cos(x)"),
    (np.tan, (-np.pi/2 + 0.1, np.pi/2 - 0.1), "Bài 7: Đạo hàm của hàm tan(x)"),
    (np.sqrt, (0.1, 4), "Bài 8: Đạo hàm của hàm căn bậc hai")
]

for f, x_range, title in functions:
    x_test = np.linspace(x_range[0], x_range[1], 3)  # Chọn 5 điểm để in kết quả
    derivatives = [calculate_derivative(f, x) for x in x_test]

    print(f"\n{title}")
    for x_val, deriv in zip(x_test, derivatives):
        print(f"f'({x_val:.3f}) ≈ {deriv:.5f}")

    plot_function_and_derivative(f, x_range, title)

```
