```python
import pandas as pd

data = pd.DataFrame({
    "Sốt": ["Cao", "Cao", "Cao", "Thấp", "Cao", "Thấp", "Thấp", "Cao", "Thấp", "Cao"],
    "Ho": ["Có", "Không", "Có", "Có", "Có", "Không", "Có", "Không", "Không", "Có"],
    "Đau họng": ["Có", "Có", "Không", "Có", "Có", "Có", "Không", "Có", "Không", "Có"],
    "Mệt mỏi": ["Có", "Không", "Có", "Có", "Không", "Không", "Có", "Có", "Không", "Có"],
    "Flu": ["Có", "Có", "Có", "Không", "Có", "Không", "Không", "Có", "Không", "Có"]
})

new_patient = {"Sốt": "Cao", "Ho": "Có", "Đau họng": "Có", "Mệt mỏi": "Không"}


P_flu = data["Flu"].value_counts(normalize=True)["Có"]
P_not_flu = 1 - P_flu

print(f"P(Flu=Có) = {P_flu:.4f}")
print(f"P(Flu=Không) = {P_not_flu:.4f}\n")

def conditional_prob(feature, value, flu_status):
    subset = data[data["Flu"] == flu_status]
    count_value = (subset[feature] == value).sum()
    total_count = len(subset)
    prob = (count_value + 1) / (total_count + 2)
    print(f"P({feature}={value} | Flu={flu_status}) = {prob:.4f}")
    return prob


P_X_given_flu = 1
P_X_given_not_flu = 1

for feature, value in new_patient.items():
    P_X_given_flu *= conditional_prob(feature, value, "Có")
    P_X_given_not_flu *= conditional_prob(feature, value, "Không")

# Tính P(C | X)
numerator_flu = P_X_given_flu * P_flu
numerator_not_flu = P_X_given_not_flu * P_not_flu

total = numerator_flu + numerator_not_flu
P_flu_given_X = numerator_flu / total
P_not_flu_given_X = numerator_not_flu / total

print("\nKết quả Naive Bayes:")
print(f"P(Flu=Có | X) = {P_flu_given_X:.4f}")
print(f"P(Flu=Không | X) = {P_not_flu_given_X:.4f}")

print("\nKết luận:", "Bệnh nhân có khả năng bị cúm." if P_flu_given_X > P_not_flu_given_X else "Bệnh nhân KHÔNG có khả năng bị cúm.")
```

```python
import pandas as pd

# Tạo DataFrame
data = pd.DataFrame({
    "Sốt": ["Cao", "Cao", "Cao", "Thấp", "Cao", "Thấp", "Thấp", "Cao", "Thấp", "Cao"],
    "Ho": ["Có", "Không", "Có", "Có", "Có", "Không", "Có", "Không", "Không", "Có"],
    "Đau họng": ["Có", "Có", "Không", "Có", "Có", "Có", "Không", "Có", "Không", "Có"],
    "Mệt mỏi": ["Có", "Không", "Có", "Có", "Không", "Không", "Có", "Có", "Không", "Có"],
    "Flu": ["Có", "Có", "Có", "Không", "Có", "Không", "Không", "Có", "Không", "Có"]
})

# Bệnh nhân mới cần dự đoán
new_patient = {"Sốt": "Cao", "Ho": "Có", "Đau họng": "Có", "Mệt mỏi": "Không"}


flu_count = 0
total_count = len(data)

print("\n [Bắt đầu vòng for 1: Tính P(Flu)]")
for i in range(total_count):
    if data["Flu"][i] == "Có":
        flu_count += 1
    print(f"  Dòng {i}: Flu = {data['Flu'][i]} → Tổng Flu='Có' = {flu_count}")
print(data["Flu"].iloc[0])
P_flu = flu_count / total_count
P_not_flu = 1 - P_flu

print(f"\n Kết quả vòng for 1:")
print(f"P(Flu=Có) = {P_flu:.4f}")
print(f"P(Flu=Không) = {P_not_flu:.4f}")

# ---  Vòng for 2: Tính P(X_i | Flu) ---
def conditional_prob(feature, value, flu_status):
    count = 0
    flu_total = 0

    print(f"\n [Bắt đầu tính P({feature}={value} | Flu={flu_status})]")

    for i in range(len(data)):  # Lặp qua từng dòng để đếm giá trị thỏa mãn
        if data["Flu"][i] == flu_status:
            flu_total += 1
            if data[feature][i] == value:
                count += 1
        print(f"  Dòng {i}: {feature}={data[feature][i]}, Flu={data['Flu'][i]} → count={count}, total={flu_total}")

    prob = count / flu_total if flu_total > 0 else 0  # Tránh chia 0
    print(f" P({feature}={value} | Flu={flu_status}) = {prob:.4f}")
    return prob


P_X_given_flu = 1
P_X_given_not_flu = 1

print("\n🔹 [Bắt đầu vòng for 3: Tính P(X | Flu)]")
for feature, value in new_patient.items():  # Lặp qua từng đặc trưng
    prob_flu = conditional_prob(feature, value, "Có")
    prob_not_flu = conditional_prob(feature, value, "Không")

    P_X_given_flu *= prob_flu
    P_X_given_not_flu *= prob_not_flu

    print(f" Sau khi nhân {feature}: P(X | Flu=Có) = {P_X_given_flu:.6f}, P(X | Flu=Không) = {P_X_given_not_flu:.6f}")

print(f"\nKết quả vòng for 3:")
print(f"P(X | Flu=Có) = {P_X_given_flu:.6f}")
print(f"P(X | Flu=Không) = {P_X_given_not_flu:.6f}")

# --- Tính P(Flu | X) ---
numerator_flu = P_X_given_flu * P_flu
numerator_not_flu = P_X_given_not_flu * P_not_flu

total = numerator_flu + numerator_not_flu
P_flu_given_X = numerator_flu / total
P_not_flu_given_X = numerator_not_flu / total

print("\n [Kết quả Naive Bayes]")
print(f"P(Flu=Có | X) = {P_flu_given_X:.4f}")
print(f"P(Flu=Không | X) = {P_not_flu_given_X:.4f}")


print("\n [Kết luận]")
if P_flu_given_X > P_not_flu_given_X:
    print("Bệnh nhân có khả năng bị cúm.")
else:
    print("Bệnh nhân KHÔNG có khả năng bị cúm.")
```
