```python
import pandas as pd

data = pd.DataFrame({
    "S·ªët": ["Cao", "Cao", "Cao", "Th·∫•p", "Cao", "Th·∫•p", "Th·∫•p", "Cao", "Th·∫•p", "Cao"],
    "Ho": ["C√≥", "Kh√¥ng", "C√≥", "C√≥", "C√≥", "Kh√¥ng", "C√≥", "Kh√¥ng", "Kh√¥ng", "C√≥"],
    "ƒêau h·ªçng": ["C√≥", "C√≥", "Kh√¥ng", "C√≥", "C√≥", "C√≥", "Kh√¥ng", "C√≥", "Kh√¥ng", "C√≥"],
    "M·ªát m·ªèi": ["C√≥", "Kh√¥ng", "C√≥", "C√≥", "Kh√¥ng", "Kh√¥ng", "C√≥", "C√≥", "Kh√¥ng", "C√≥"],
    "Flu": ["C√≥", "C√≥", "C√≥", "Kh√¥ng", "C√≥", "Kh√¥ng", "Kh√¥ng", "C√≥", "Kh√¥ng", "C√≥"]
})

new_patient = {"S·ªët": "Cao", "Ho": "C√≥", "ƒêau h·ªçng": "C√≥", "M·ªát m·ªèi": "Kh√¥ng"}


P_flu = data["Flu"].value_counts(normalize=True)["C√≥"]
P_not_flu = 1 - P_flu

print(f"P(Flu=C√≥) = {P_flu:.4f}")
print(f"P(Flu=Kh√¥ng) = {P_not_flu:.4f}\n")

def conditional_prob(feature, value, flu_status):
    subset = data[data["Flu"] == flu_status]
    count_value = (subset[feature] == value).sum()
    total_count = len(subset)
    prob = (count_value + 1) / (total_count + 2)
    print(f"P({feature}={value} | Flu={flu_status}) = {prob:.4f}")
    return prob


P_X_given_flu = 1
P_X_given_not_flu = 1

for feature, value in new_patient.items():
    P_X_given_flu *= conditional_prob(feature, value, "C√≥")
    P_X_given_not_flu *= conditional_prob(feature, value, "Kh√¥ng")

# T√≠nh P(C | X)
numerator_flu = P_X_given_flu * P_flu
numerator_not_flu = P_X_given_not_flu * P_not_flu

total = numerator_flu + numerator_not_flu
P_flu_given_X = numerator_flu / total
P_not_flu_given_X = numerator_not_flu / total

print("\nK·∫øt qu·∫£ Naive Bayes:")
print(f"P(Flu=C√≥ | X) = {P_flu_given_X:.4f}")
print(f"P(Flu=Kh√¥ng | X) = {P_not_flu_given_X:.4f}")

print("\nK·∫øt lu·∫≠n:", "B·ªánh nh√¢n c√≥ kh·∫£ nƒÉng b·ªã c√∫m." if P_flu_given_X > P_not_flu_given_X else "B·ªánh nh√¢n KH√îNG c√≥ kh·∫£ nƒÉng b·ªã c√∫m.")
```

```python
import pandas as pd

# T·∫°o DataFrame
data = pd.DataFrame({
    "S·ªët": ["Cao", "Cao", "Cao", "Th·∫•p", "Cao", "Th·∫•p", "Th·∫•p", "Cao", "Th·∫•p", "Cao"],
    "Ho": ["C√≥", "Kh√¥ng", "C√≥", "C√≥", "C√≥", "Kh√¥ng", "C√≥", "Kh√¥ng", "Kh√¥ng", "C√≥"],
    "ƒêau h·ªçng": ["C√≥", "C√≥", "Kh√¥ng", "C√≥", "C√≥", "C√≥", "Kh√¥ng", "C√≥", "Kh√¥ng", "C√≥"],
    "M·ªát m·ªèi": ["C√≥", "Kh√¥ng", "C√≥", "C√≥", "Kh√¥ng", "Kh√¥ng", "C√≥", "C√≥", "Kh√¥ng", "C√≥"],
    "Flu": ["C√≥", "C√≥", "C√≥", "Kh√¥ng", "C√≥", "Kh√¥ng", "Kh√¥ng", "C√≥", "Kh√¥ng", "C√≥"]
})

# B·ªánh nh√¢n m·ªõi c·∫ßn d·ª± ƒëo√°n
new_patient = {"S·ªët": "Cao", "Ho": "C√≥", "ƒêau h·ªçng": "C√≥", "M·ªát m·ªèi": "Kh√¥ng"}


flu_count = 0
total_count = len(data)

print("\n [B·∫Øt ƒë·∫ßu v√≤ng for 1: T√≠nh P(Flu)]")
for i in range(total_count):
    if data["Flu"][i] == "C√≥":
        flu_count += 1
    print(f"  D√≤ng {i}: Flu = {data['Flu'][i]} ‚Üí T·ªïng Flu='C√≥' = {flu_count}")
print(data["Flu"].iloc[0])
P_flu = flu_count / total_count
P_not_flu = 1 - P_flu

print(f"\n K·∫øt qu·∫£ v√≤ng for 1:")
print(f"P(Flu=C√≥) = {P_flu:.4f}")
print(f"P(Flu=Kh√¥ng) = {P_not_flu:.4f}")

# ---  V√≤ng for 2: T√≠nh P(X_i | Flu) ---
def conditional_prob(feature, value, flu_status):
    count = 0
    flu_total = 0

    print(f"\n [B·∫Øt ƒë·∫ßu t√≠nh P({feature}={value} | Flu={flu_status})]")

    for i in range(len(data)):  # L·∫∑p qua t·ª´ng d√≤ng ƒë·ªÉ ƒë·∫øm gi√° tr·ªã th·ªèa m√£n
        if data["Flu"][i] == flu_status:
            flu_total += 1
            if data[feature][i] == value:
                count += 1
        print(f"  D√≤ng {i}: {feature}={data[feature][i]}, Flu={data['Flu'][i]} ‚Üí count={count}, total={flu_total}")

    prob = count / flu_total if flu_total > 0 else 0  # Tr√°nh chia 0
    print(f" P({feature}={value} | Flu={flu_status}) = {prob:.4f}")
    return prob


P_X_given_flu = 1
P_X_given_not_flu = 1

print("\nüîπ [B·∫Øt ƒë·∫ßu v√≤ng for 3: T√≠nh P(X | Flu)]")
for feature, value in new_patient.items():  # L·∫∑p qua t·ª´ng ƒë·∫∑c tr∆∞ng
    prob_flu = conditional_prob(feature, value, "C√≥")
    prob_not_flu = conditional_prob(feature, value, "Kh√¥ng")

    P_X_given_flu *= prob_flu
    P_X_given_not_flu *= prob_not_flu

    print(f" Sau khi nh√¢n {feature}: P(X | Flu=C√≥) = {P_X_given_flu:.6f}, P(X | Flu=Kh√¥ng) = {P_X_given_not_flu:.6f}")

print(f"\nK·∫øt qu·∫£ v√≤ng for 3:")
print(f"P(X | Flu=C√≥) = {P_X_given_flu:.6f}")
print(f"P(X | Flu=Kh√¥ng) = {P_X_given_not_flu:.6f}")

# --- T√≠nh P(Flu | X) ---
numerator_flu = P_X_given_flu * P_flu
numerator_not_flu = P_X_given_not_flu * P_not_flu

total = numerator_flu + numerator_not_flu
P_flu_given_X = numerator_flu / total
P_not_flu_given_X = numerator_not_flu / total

print("\n [K·∫øt qu·∫£ Naive Bayes]")
print(f"P(Flu=C√≥ | X) = {P_flu_given_X:.4f}")
print(f"P(Flu=Kh√¥ng | X) = {P_not_flu_given_X:.4f}")


print("\n [K·∫øt lu·∫≠n]")
if P_flu_given_X > P_not_flu_given_X:
    print("B·ªánh nh√¢n c√≥ kh·∫£ nƒÉng b·ªã c√∫m.")
else:
    print("B·ªánh nh√¢n KH√îNG c√≥ kh·∫£ nƒÉng b·ªã c√∫m.")
```
