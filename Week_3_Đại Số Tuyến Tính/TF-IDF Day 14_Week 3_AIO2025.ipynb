```python
import numpy as np
documents=["Tôi thích học AI","AI là trí tuệ nhân tạo","AGI là siêu trí tuệ nhân tạo"]
words=set()
#chạy vòng lặp cho từng doc trong document rồi tách từng từ ra = splti
for doc in documents:
  words.update(doc.split())
words=list(words)
print(words)
```

```python
def compute_tf(doc):
  tf = {} #khởi tạo tf
  total_words = len(doc.split())
#chạy vòng lặp cho từng từ trong document đã split
  for word in doc.split():
    #nếu có trong tf thì điền + dồn
    if word in tf:
      tf[word] += 1
      #nếu chưa có thì điền 1
    else:
      tf[word]  = 1
      #chạy vòng lặp trong tf
  for word in tf:
    tf[word]  /= total_words
  return tf

compute_tf(documents[2])
```

```python
def compute_idf(documents):
  idf={} #khởi tạo idf
  #chạy vòng lặp cho từng từ trong ds words
  for word in words:
    count = 0
    #chạy vòng lặp từng doc trong documents
    for doc in documents:
      #nếu từ xuất hiện trong doc tăng biến lên 1
      if word in doc.split():
        count += 1
        #tính idf theo công thức
    idf[word] = np.log(len(documents)/(count+1))
  return idf
idf_values = compute_idf(documents)
print(idf_values)
```

```python
def compute_tf_idf(documents):
  idf = compute_idf(documents)
  tf_idf_documents = [] #khởi tạo tf_idf
#chạy vòng lặp cho từng doc trong documents
  for doc in documents:
    #tính tf cho từng doc
    tf = compute_tf(doc)
    #tính tf_idf cho từng từ
    tf_idf = {word: tf[word] * idf[word] for word in tf}
    #lưu kết quả cho ds
    tf_idf_documents.append(tf_idf)

  return tf_idf_documents
```

```python
tf_idf_documents = compute_tf_idf(documents)
#chạy vòng lặp trên index tất cả documents
for i in range(len(documents)):
    print(f"Document {i+1}: {documents[i]}")

    tf_idf = tf_idf_documents[i]
    #chạy vòng lặp cho từng từ trong tf_idf
    for word in tf_idf:
        value = tf_idf[word]
        print(word, ":", value)
```
