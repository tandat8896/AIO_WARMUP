```python
#câu 1
A=[]
for i in range(3):
  row = []
  for j in range (1,4):
    row.append(i*3+j)
  A.append(row)
print("A=",A)

B = [[2, 4],
     [1, 3]]
print("B=",B)

C = [[1,1,1],
     [0,0,0],
     [1,1,1]]
print("C=",C)
```

```python
def convolution(A,B):
  n_rows = len(A)-len(B)+1 #số hàng mà kernal có thể trượt trong A
  n_cols = len(A[0])-len(B[0])+1 #số cột mà kernal có thể trượt trong A

  if n_rows <= 0 or n_cols <= 0:
    return []
  result=[]
#chạy vòng lặp cho từng hàng trong vùng có thể trượt trong A
  for i in range(n_rows):
    row = []
    #chạy từng cột trong từng hàng mà kernal có thể trượt trong A
    for j in range(n_cols):
      sum = 0
      for k in range(len(B)): #lấy từng hàng trong B
        for l in range(len(B[0])): #lấy từng cột trong B
          sum += A[i+k][j+l]*B[k][l] #tổng của B và vùng cần trượt
      row.append(sum)
    result.append(row)
    print(f"i={i}, j={j}, k={k}, l={l}, value={A[i + k][j + l]}")

  return result
print(result)

output=convolution(A,B)
print(output)
output1=convolution(A,C)
print(output1)
```

```python
#câu 2
A=[[0,0,0],
   [0,4,0],
   [0,1,0]]
print("A=",A)
B=[[1,1],
   [1,1]
]
print("B=",B)
C=[[0,1,0],
   [0,1,0],
   [0,1,0]]
print("C=",C)

def convolution_pad(A,B,Pad):
  n_rows = len(A)
  n_cols = len(A[0])
  padded_rows=n_rows+2*Pad
  padded_cols=n_cols+2*Pad
  padded_matrix = [[0 for _ in range(padded_cols)] for _ in range(padded_rows)]
  for i in range(n_rows):
    for j in range(n_cols):
      padded_matrix[i+Pad][j+Pad]=A[i][j]
  return convolution(padded_matrix,B)

#cau1:
for row in convolution_pad(A,B,1):
  print(row)
#cau2:
for row in convolution_pad(A,C,1):
  print(row)

```

```python
#bài 3
A=[[0,0,0,4],
   [0,4,0,2],
   [0,1,0,2],
   [0,3,0,2]]

def max_average_pooling(A, pool_size):
    n_rows = len(A)
    n_cols = len(A[0])
    #số hàng của ma trận output
    out_rows = n_rows // pool_size
    #số cột của ma trận ouput
    out_cols = n_cols // pool_size
    #khởi tạo vòng lặp cho các hàng và các cột = 0
    output = [[0 for _ in range(out_cols)] for _ in range(out_rows)]
    output2 = [[0 for _ in range(out_cols)] for _ in range(out_rows)]
    #lấy các hàng trong ma trận A với bước nhảy= pool
    for i in range(0, n_rows, pool_size):
      #lấy các cột trong hàng của ma trận A với pool_size
        for j in range(0, n_cols, pool_size):
          #tạo ma trận pool rỗng
            pool = []
            #chạy vòng lặp trên từng hàng ma trận con pool ra
            for k in range(pool_size):
              #chạy vòng lặp trên từng cột trên hàng của ma trận con
                for l in range(pool_size):
                  #điều kiện xem ma trận có còn nằm trong phạm vi A
                    if i + k <= n_rows and j + l <= n_cols:
                      #điền kết quả vào ma trận pool
                        pool.append(A[i + k][j + l])
                        #định nghĩa vị trí pool cho output
                        #output = lấy số lớn nhất trên ma trận pool trên
                        row_index=i//pool_size
                        col_index=j//pool_size
                        output[row_index][col_index] = max(pool)
                        output2[row_index][col_index] = sum(pool) / len(pool)
                        print(f"i={i}, j={j}, k={k}, l={l}, value={A[i + k][j + l]}")

    return output , output2


result_max, result_avg = max_average_pooling(A, 2)
print(result_max)
print(result_avg)
```
