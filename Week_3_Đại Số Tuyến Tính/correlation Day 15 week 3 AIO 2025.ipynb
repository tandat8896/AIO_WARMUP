```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/advertising.csv')
```

```python
x = data['TV']
y = data['Radio']
```

```python
def correlation(x,y):
  n=len(x)
  if len(x) != len(y):
    return None

  sum_x = np.sum(x)
  mean_x = sum_x/len(x)
  sum_y = np.sum(y)
  mean_y = sum_y/len(y)
  numerator = denominator_x = denominator_y =0
  for i in range(n):
    diff_x = x[i] - mean_x
    diff_y = y[i] - mean_y
    numerator += (diff_x*diff_y)
    denominator_x += diff_x ** 2
    denominator_y += diff_y ** 2

    denominator = np.sqrt(denominator_x * denominator_y)

  if denominator !=0:
    correlation = numerator/denominator
  else:
    correlation = 0
  return correlation

output = correlation(x,y)
print(output)
```

```python
lst_data = list(data)
print(lst_data)
```

```python
for i in lst_data:
    for j in lst_data:
        output = correlation(data[i], data[j])
        print(f"Correlation between {i} and {j}: {output}")
```

```python

lst_data = list(data)
output = []
for i in lst_data:
  row = []
  for j in lst_data:
    row.append(correlation(data[i],data[j]))
  output.append(row)
output = np.array(output)
print(output)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(output, annot=True, fmt =".2f",linewidth=5)
plt.show()
```
