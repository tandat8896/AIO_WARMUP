```python
possible_moves = [
    (2, 1), (1, 2), (-1, 2), (-2, 1),
    (-2, -1), (-1, -2), (1, -2), (2, -1)
]

def procedure_knightts_tour(x, y, move_count, N, board, possible_moves):
    if move_count == N * N:
        return True

    for delta_x, delta_y in possible_moves:
        new_x = x + delta_x
        new_y = y + delta_y
        if 0 <= new_x < N and 0 <= new_y < N and board[new_x][new_y] == -1: #kiểm tra xem new_x và new_y có nằm trong bàn cờ không và nước đi đó đã đi chưa board[new_x][new_y] chưa đi =-1
            board[new_x][new_y] = move_count
            if procedure_knightts_tour(new_x, new_y, move_count + 1, N, board, possible_moves): #nếu đi được hết bàn cờ thì return true dùng backtracking đệ quy nếu ngược lại
                return True
            board[new_x][new_y] = -1  #backtracking cho trường hợp trên nếu không tìm được thì return false
    return False

def function_solve_knights_tour(N, start_x, start_y):
    board = [[-1 for _ in range(N)] for _ in range(N)]  #khởi tạo bàn cờ với các giá trị -1 trong  hàng và trong cột
    board[start_x][start_y] = 0   #ban đầu ở vị trí 0,0

    if procedure_knightts_tour(start_x, start_y, 1, N, board, possible_moves): #thử tìm trên 1 lần đi
        print(board)
    else:
        print("No solution")


function_solve_knights_tour(5, 0, 0)

```
