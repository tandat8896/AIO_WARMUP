```python
import cv2 
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread('nature.jpg')
```

```python
cv2.imshow('Windows Name', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('New_nature.jpg',image)
```

```python
image_bgr = cv2.imread('image.jpg')
image_rgb = cv2.cvtColor(image_bgr,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
image_brg = cv2.imread('Image_6385.jpg')
image_hsv = cv2.cvtColor(image_bgr,cv2.COLOR_BGR2HSV)
plt.imshow(image_hsv)
plt.axis('off')
plt.show()
```

```python
image_old = cv2.imread('image.jpg')
image_new = cv2.cvtColor(image_old , cv2.COLOR_BGR2GRAY)
save_image = cv2.imwrite(filename='NewGrayScal.jpg',img=image_new)
image_gray = cv2.imread('NewGrayScal.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image_gray,cmap='gray')
plt.axis('Off')
plt.show()
```

```python
combine = cv2.hconcat([image_rgb,image_hsv])
cv2.imshow('comparision',combine)
cv2.waitKey(0)
cv2.destroyAllWindows
```
