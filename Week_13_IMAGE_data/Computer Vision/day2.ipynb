```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Image_6385.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6, 6))
plt.imshow(img)
plt.axis('off')
plt.show()


# Convert the image to different color spaces
img_bgr = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
img_hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# Create a grid plot to display the images
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
axes = axes.ravel()

# Display the original RGB image
axes[0].imshow(img)
axes[0].set_title('RGB')
axes[0].axis('off')

# Display the BGR image
axes[1].imshow(img_bgr)
axes[1].set_title('BGR')
axes[1].axis('off')

# Display the HSV image
axes[2].imshow(img_hsv)
axes[2].set_title('HSV')
axes[2].axis('off')

# Display the HSB image
axes[3].imshow(img_hls)
axes[3].set_title('HSL')
axes[3].axis('off')

# Display the Grayscale image
axes[4].imshow(img_gray, cmap='gray')
axes[4].set_title('Grayscale')
axes[4].axis('off')

# Hide the last subplot (empty)
axes[5].axis('off')

plt.tight_layout()
plt.show()



```
