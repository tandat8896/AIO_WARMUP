```python
# increase the brightness using opencv function
import cv2
import numpy as np

def increase_brightness(image_path, beta_value, output_path):
    # Load the image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Increase brightness by adding beta_value to all pixel intensities
    # output=input+
    brightened_image = cv2.convertScaleAbs(image, alpha=1, beta=beta_value)

    # Save the brightened image
    cv2.imwrite(output_path, brightened_image)

# Parameters
input_image_path = "/content/nature.jpg"  # Replace with your grayscale image file path
output_image_path = "/content/nature_gs_inreasing.jpg"  # File path to save the brightened image
brightness_increase = 50  # Adjust this value to change brightness (higher value = brighter)

# Call the function
increase_brightness(input_image_path, brightness_increase, output_image_path)
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def increase_brightness_customize(image_path, brightness_value, output_path):
    # Step 1: Read image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Step 2: Convert the image to a 1D list (flattened array)
    flattened_image = image.flatten()

    # Step 3: Increase the brightness in the 1D list using a for loop
    brightened_1d = []
    for pixel_value in flattened_image:
        # Increase brightness and ensure pixel values are between 0 and 255
        brightened_pixel = pixel_value + brightness_value
        brightened_1d.append(np.clip(brightened_pixel, 0, 255))

    # Step 4: Reshape the 1D list back to the original grayscale image shape
    brightened_image = np.array(brightened_1d).reshape(image.shape)

    # Save the brightened image
    cv2.imwrite(output_path, brightened_image)

    return brightened_image

input_image_path = "/content/nature.jpg"  # Replace with your grayscale image file path
output_image_path = "/content/nature_gs_inreasing_customize.jpg"  # File path to save the brightened image
brightness_increase = 50  # Adjust this value to change brightness (higher value = brighter)

# Call the function
increase_brightness_customize(input_image_path, brightness_increase, output_image_path)
```

```python
def increase_brightness_customizev2(image_path, brightness_value, output_path):
  img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
  img = img.astype(float)

  img = img + brightness_value
  img = np.where(img<0., 0., img)
  img = np.where(img>255., 255., img)

  img = img.astype(np.uint8)
  cv2.imwrite(output_path, img)

input_image_path = "/content/nature.jpg"  # Replace with your grayscale image file path
output_image_path = "/content/nature_gs_inreasing_customize_v2.jpg"  # File path to save the brightened image
brightness_increase = 50  # Adjust this value to change brightness (higher value = brighter)
increase_brightness_customizev2(input_image_path, brightness_increase, output_image_path)
```
