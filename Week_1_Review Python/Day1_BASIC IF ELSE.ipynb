```python
def calculate_can_chi_calendar(year):
    can_list = ["Canh", "Tân", "Nhâm", "Quý", "Giáp", "Ất", "Bính", "Đinh", "Mậu", "Kỷ"]
    chi_list = ["Thân", "Dậu", "Tuất", "Hợi", "Tý", "Sửu", "Dần", "Mão", "Thìn", "Tỵ", "Ngọ", "Mùi"]

    N = len(can_list)
    N1 = len(chi_list)
    for index in range(N):
        if year % 10 == index:
            can = can_list[index]
            break

    for index in range(N1):
        if year % 12 == index:
            chi = chi_list[index]
            break

    result = f"{can} {chi}"
    return result
```

```python
print(calculate_can_chi_calendar(2025))
```

```python
print(calculate_can_chi_calendar(2024))
print(calculate_can_chi_calendar(1997))
```

```python
def relu(x):
    if x > 0:
        return x
    else:
        return 0
```

```python

print(relu(3))


print(relu(-2))

print(relu(0))

```

```python
def leaky_relu(x, alpha=0.01):
    if x > 0:
        return x
    else:
        return alpha * x
```

```python
print(leaky_relu(3))
print(leaky_relu(-2))
print(leaky_relu(0))
```

```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))


print(sigmoid(0))
print(sigmoid(2))
print(sigmoid(-2))
```

```python


def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))


print(tanh(0))
print(tanh(2))
print(tanh(150))
```

```python


def elu(x, alpha=1.0):
    return x if x > 0 else alpha * (math.exp(x) - 1)


print(elu(3))
print(elu(-1))
print(elu(0))
```
