```python
import random

def random_number_with_condition(total):
    count = 0
    random.seed(0)

    while True:
        a = random.randint(1, 20)
        b = random.randint(1, 20)
        count += 1
        if a + b == total:
            break

    return count


print(f"Test case với tổng 40: {random_number_with_condition(40)}")
print(f"Test case với tổng 20: {random_number_with_condition(20)}")
print(f"Test case với tổng 35: {random_number_with_condition(35)}")
```
