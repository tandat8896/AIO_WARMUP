```python
import math

def compute_interest(money, period):
    for _ in range(period):
        money *= 1 + (1 / period)
    return money

print(compute_interest(1, 12))
print(compute_interest(1, 365))
print(compute_interest(1, 730))
```

```python
def is_prime_eratosthenes(n):
    if n < 2:
        return False

    prime = [True] * (n + 1)
    prime[0] = prime[1] = False  # 0 và 1 không phải số nguyên tố

    for i in range(2, int(math.sqrt(n)) + 1):
        if prime[i]:
            for j in range(i * i, n + 1, i):  # Đánh dấu các bội số của i
                prime[j] = False

    return prime[n]


n = 19
is_prime_result = is_prime_eratosthenes(n)
is_prime_result
```
