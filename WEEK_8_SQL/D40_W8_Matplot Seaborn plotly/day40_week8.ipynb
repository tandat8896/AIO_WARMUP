```python
df = pd.read_csv('menu.csv')
print(df.columns)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))


sns.histplot(df, x="Total Fat", ax=axes[0], bins=20)
axes[0].set_xlabel("Total Fat")
axes[0].set_ylabel("Count")


sns.histplot(df, x="Total Fat", ax=axes[1], bins=20, stat="density", kde=True)
axes[1].set_xlabel("Total Fat")
axes[1].set_ylabel("Density")

plt.tight_layout()
plt.show()
```

```python
corr_value = df["Total Fat"].corr(df["Calories"])
print("Hệ số tương quan giữa Total Fat và Calories:", corr_value)


sns.set_style({"axes.grid": False})
sns.regplot(x="Total Fat", y="Calories", data=df)


plt.xlabel("Total Fat")
plt.ylabel("Calories")
plt.show()
```

```python
sns.set_theme(style="whitegrid")
g = sns.jointplot(
    data=df,
    x="Total Fat",
    y="Calories",
    kind="reg",
    height=6,
    marginal_kws=dict(bins=20, fill=True)
)


g.fig.suptitle("Phân phối và tương quan Total Fat - Calories", y=1.02)

plt.show()
```

```python
sns.boxplot(data=df, y="Calories")
plt.show()
```

```python
Q1 = df["Sodium"].quantile(0.25)
Q3 = df["Sodium"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df["Sodium"] >= lower_bound) & (df["Sodium"] <= upper_bound)]
```

```python
sns.boxplot(data=df, y="Calories")
plt.show()
```

```python
result = df.groupby("Category")["Item"].count().reset_index(name="Item Counts")
print(result)
```

```python
num_categories = result["Category"].nunique()


colors = sns.color_palette("Set2", n_colors=num_categories)

plt.figure(figsize=(10, 6))
sns.barplot(
    data=result,
    x="Category",
    y="Item Counts",
    hue="Category",
    palette=colors,
    legend=False
)
plt.title("Biểu đồ số lượng Items có trong Category")
plt.xlabel("Category")
plt.ylabel("Item Counts")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
