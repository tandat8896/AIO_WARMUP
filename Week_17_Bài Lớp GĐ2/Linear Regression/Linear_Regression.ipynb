```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
def f(x):
    return x**2 

def f_prime(x):
    return 2*x 
```

```python

def gradient_decent(x0, lr , max_iter = 100 , epsilon = 1e-6):
    x= x0
    history = [(x,f(x))]
    for i in range(max_iter):
        grad = f_prime(x)
        x_new = x - lr * grad
        history.append((x_new , f(x_new)))
        if i % 10 == 0 :
            print(f"Iteration {i}: x= {x_new:.6f}, f(x) = {f(x_new):.6f}")

        if abs(x_new - x ) < epsilon:
            print(f"Convergence at Iteration {i+1}:x= {x_new:.6f}, f(x) = {f(x_new):.6f}")
            break
        x=x_new
    return np.array(history)
```

```python
# Run GD
x0 = 10.0
lr = 0.1
history = gradient_decent(x0, lr)
```

```python
#create blue graph
x_vals = np.linspace(-11, 11,400)
y_vals = f(x_vals)
plt.plot(x_vals,y_vals , label = '$f(x) = x^2$', color = 'blue')

#scatter history mean the value of x in the every moving step 
plt.scatter(history[:,0],history[:,1], color='red',label = 'GD steps', s=50 , zorder =5)
plt.plot(history[:,0],history[:,1], color = 'red', alpha =0.5 , linestyle='--')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent on $f(x) = x^2$ ($\\eta = 0.1$)')
plt.legend()
plt.grid(True)
plt.show()
```
