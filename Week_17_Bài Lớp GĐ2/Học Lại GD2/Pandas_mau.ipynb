# Data Transformation & Feature Engineering

## apply()

```python
df = df.drop(columns=['grade.average'])
```

```python
# Create average grade column
df['average_grade'] = df[['english.grade', 'math.grade', 'sciences.grade', 'language.grade']].mean(axis=1)
df['average_grade'].head()
```

```python
df['performance_category'] = df['average_grade'].apply(
    lambda x: 'Excellent' if x >= 3.6 else ('Good' if x >= 3.2 else 'Needs Improvement')
)
```

```python
# Apply function to create performance category
def performance_category(row):
    if row['average_grade'] >= 3.6:
        return 'Excellent'
    elif row['average_grade'] >= 3.2:
        return 'Good'
    else:
        return 'Needs Improvement'

df['performance_category'] = df.apply(performance_category, axis=1)
df.head()
```

## map()

```python
# Return original value of gender
df['gender'] = df['gender'].str.replace('female', 'F').replace('male', 'M')
```

```python
# Map gender codes to binary numeric values
gender_map = {'M': '0', 'F': '1'}
df['gender'] = df['gender'].map(gender_map)
```

```python
# Map letter grades
coverletter_rating = {4.0:'A', 3.0:'B', 2.0:'C', 1.0:'D', 0.0:'F'}
df['coverletter_rating_alphabet'] = df['coverletter.rating'].map(coverletter_rating)
```

```python
# Using map with a function
df['name.length'] = df['name'].map(len)
```

```python
df.head()
```

## Binning & Discretization

```python
# Bin ages into categories
age_bins = [19, 20, 21, 22, 26]
age_labels = ['fresher', 'junior', 'middle', 'senior']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)
df['age_group'].head()
```

# Aggregation & Grouping

```python
# Group by nationality and calculate mean grades
nationality_stats = df.groupby('nationality')[['english.grade', 'math.grade']].mean()
nationality_stats.head()
```

```python
studen_grade_levels = df.groupby('coverletter_rating_alphabet')[['name']].count()
studen_grade_levels.head()
```

## Pivot table

```python
df = pd.read_csv("/content/drive/MyDrive/Data/student.csv")
```

```python
# Create average grade column
df['average_grade'] = df[['english.grade', 'math.grade', 'sciences.grade', 'language.grade']].mean(axis=1)
df['average_grade'].head()
```

```python
# Pivot table: Average grade by nationality and gender
pivot_grade = pd.pivot_table(df,
                            values='average_grade',
                            index='nationality',
                            columns='gender',
                            aggfunc=lambda x: round(x.mean(), 2),
                            fill_value=0)

print("\nPivot Table - Average Grades:")
pivot_grade
```
