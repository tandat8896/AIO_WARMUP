```python
import psutil

# Lấy thông tin bộ nhớ hệ thống
memory_info = psutil.virtual_memory()

# In ra tổng bộ nhớ, bộ nhớ đã sử dụng và bộ nhớ còn lại
print(f"Tổng bộ nhớ: {memory_info.total / (1024 ** 3):.2f} GB")
print(f"Bộ nhớ đã sử dụng: {memory_info.used / (1024 ** 3):.2f} GB")
print(f"Bộ nhớ còn lại: {memory_info.available / (1024 ** 3):.2f} GB")
print(f"Phần trăm bộ nhớ đã sử dụng: {memory_info.percent}%")
```

```python


# Lấy danh sách tất cả các tiến trình đang chạy
processes = [p.info for p in psutil.process_iter(['pid', 'name', 'memory_info'])]

# Sắp xếp các tiến trình theo bộ nhớ đã sử dụng (tính bằng MB)
processes_sorted = sorted(processes, key=lambda x: x['memory_info'].rss, reverse=True)

# In ra thông tin các tiến trình sử dụng nhiều bộ nhớ nhất
for process in processes_sorted[:5]:
    print(f"PID: {process['pid']}, Tên: {process['name']}, Bộ nhớ đã sử dụng: {process['memory_info'].rss / (1024 ** 2):.2f} MB")
```
