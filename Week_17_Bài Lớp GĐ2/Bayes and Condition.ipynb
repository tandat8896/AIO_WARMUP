```python
import math

class Probability:
    def __init__(self, mean=None, std=None):
        self.mean = mean
        self.std = std

    def basic_probability(self, event, total):
        """Tính xác suất cơ bản P(A) = event / total"""
        return event / total if total > 0 else 0

    def conditional_probability(self, joint, given):
        """Tính xác suất có điều kiện P(A | B) = P(A ∩ B) / P(B)"""
        return joint / given if given > 0 else 0

    def bayes_probability(self, prior, likelihood, marginal):
        """Tính xác suất Bayes: P(A | B) = (P(B | A) * P(A)) / P(B)"""
        return (likelihood * prior) / marginal if marginal > 0 else None  # Trả về None nếu không thể tính


    def factorial(self, n):
        """Tính giai thừa n!"""
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result
```

```python
import math

class NormalDistribution:
    def __init__(self, mean=0, std=1):
        self.mean = mean
        self.std = std

    def standardize(self, x):
        return (x - self.mean) / self.std

    def taylor_series_cdf(self, x, terms=10):
        pi = math.pi
        result = 0
        for n in range(terms):
            term = ((-1) ** n) * (x ** (2 * n + 1)) / ((2 * n + 1) * math.factorial(n) * (2 ** n))
            result += term
        return 0.5 + (result / math.sqrt(2 * pi))

    def simpsons_rule_cdf(self, x, n=1000):
        def normal_pdf(x):
            return math.exp(-x**2 / 2) / math.sqrt(2 * math.pi)

        a, b = -10, x  # Giới hạn tích phân
        h = (b - a) / n
        result = normal_pdf(a) + normal_pdf(b)
        for i in range(1, n, 2):
            result += 4 * normal_pdf(a + i * h)
        for i in range(2, n-1, 2):
            result += 2 * normal_pdf(a + i * h)
        return (h / 3) * result

    def gauss_hermite_cdf(self, x):
        return 0.5 * (1 + math.erf(x / math.sqrt(2)))

    def probability(self, x, interval=None):
        if interval:
            lower, upper = self.standardize(interval[0]), self.standardize(interval[1])
            taylor = self.taylor_series_cdf(upper) - self.taylor_series_cdf(lower)
            simpson = self.simpsons_rule_cdf(upper) - self.simpsons_rule_cdf(lower)
            gauss = self.gauss_hermite_cdf(upper) - self.gauss_hermite_cdf(lower)
        else:
            x = self.standardize(x)
            taylor = self.taylor_series_cdf(x)
            simpson = self.simpsons_rule_cdf(x)
            gauss = self.gauss_hermite_cdf(x)

        return taylor, simpson, gauss

# Danh sách bài toán
cases = [
    (1, 'P(X > 1)'),
    (115, 'P(85 ≤ X ≤ 115)', 100, 15),
    (-2, 'P(X < -2)'),
    (90, 'P(X > 90)', 75, 10),
    (75, 'P(65 ≤ X ≤ 75)', 70, 8)
]

for case in cases:
    if len(case) == 2:  # Trường hợp chuẩn tắc (mean=0, std=1)
        x, label = case
        norm = NormalDistribution()
        taylor, simpson, gauss = norm.probability(x)
    else:  # Trường hợp tổng quát
        x, label, mean, std = case
        norm = NormalDistribution(mean, std)
        if '≤ X ≤' in label:
            taylor, simpson, gauss = norm.probability(x, (85, 115))
        else:
            taylor, simpson, gauss = norm.probability(x)

    if '>' in label:
        taylor = 1 - taylor
        simpson = 1 - simpson
        gauss = 1 - gauss

    print(f"{label}: Taylor={taylor:.4f}, Simpson={simpson:.4f}, Gauss-Hermite={gauss:.4f}")
```
