```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
print(tips)
```

```python
sns.histplot(tips['total_bill'], kde=True, bins=20,edgecolor='r')
```

```python
sns.violinplot(data=tips, x='time', y='total_bill',color='r')
plt.title("violinplot")
```

```python
diamonds = sns.load_dataset("diamonds")
```

```python
diamonds.head()
```

```python
sns.scatterplot(data=diamonds, x="carat", y="price", hue="cut")
```

```python
sns.lmplot(data=diamonds, x="carat", y="price", hue="cut", aspect=1.5)
```

```python
plt.figure(figsize=(12, 6))

sns.lmplot(
    data=diamonds,           # Dataset chứa các cột 'carat', 'price', 'cut', v.v.
    x='carat',               # Trục hoành: trọng lượng viên kim cương
    y='price',               # Trục tung: giá tiền
    hue='cut',               # Tô màu các điểm và đường hồi quy theo 'cut'
    col='cut',               # Tạo subplot riêng cho từng giá trị trong cột 'cut'
    col_wrap=3,              # Sắp xếp các subplot theo dạng lưới với tối đa 3 cột mỗi hàng
    scatter_kws={
        's': diamonds['depth'].mean()*2,  # Kích thước điểm scatter theo độ sâu trung bình
        'alpha': 0.6                      # Độ trong suốt của điểm scatter
    },
    line_kws={
        'lw': 2                           # Độ dày của đường hồi quy
    },
    height=4,                 # Chiều cao mỗi subplot (inch)
    aspect=1.2                # Tỉ lệ chiều rộng/chiều cao của mỗi subplot
)


plt.suptitle('Mối quan hệ Carat - Price theo chất lượng kim cương', y=1.02)
plt.tight_layout()
plt.show()
```
