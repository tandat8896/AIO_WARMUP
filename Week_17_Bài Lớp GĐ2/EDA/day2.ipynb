```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
print(tips)
```

```python
sns.histplot(tips['total_bill'], kde=True, bins=20,edgecolor='r')
```

```python
sns.violinplot(data=tips, x='time', y='total_bill',color='r')
plt.title("violinplot")
```

```python
diamonds = sns.load_dataset("diamonds")
```

```python
diamonds.head()
```

```python
sns.scatterplot(data=diamonds, x="carat", y="price", hue="cut")
```

```python
sns.lmplot(data=diamonds, x="carat", y="price", hue="cut", aspect=1.5)
```

```python
plt.figure(figsize=(12, 6))

sns.lmplot(
    data=diamonds,           # Dataset cha cc ct 'carat', 'price', 'cut', v.v.
    x='carat',               # Trc honh: trng lng vin kim cng
    y='price',               # Trc tung: gi tin
    hue='cut',               # T mu cc im v ng hi quy theo 'cut'
    col='cut',               # To subplot ring cho tng gi tr trong ct 'cut'
    col_wrap=3,              # Sp xp cc subplot theo dng li vi ti a 3 ct mi hng
    scatter_kws={
        's': diamonds['depth'].mean()*2,  # Kch thc im scatter theo  su trung bnh
        'alpha': 0.6                      #  trong sut ca im scatter
    },
    line_kws={
        'lw': 2                           #  dy ca ng hi quy
    },
    height=4,                 # Chiu cao mi subplot (inch)
    aspect=1.2                # T l chiu rng/chiu cao ca mi subplot
)


plt.suptitle('Mi quan h Carat - Price theo cht lng kim cng', y=1.02)
plt.tight_layout()
plt.show()
```
