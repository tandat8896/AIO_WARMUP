```python
x = np.linspace(start=-10,stop=10,num=1000)
y = x**2
fig,ax = plt.subplots()
ax.plot(x,y,label='y=x^2', color= 'r')
ax.set_title("Đồ Thị hàm số y=x^2")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True)
ax.legend()
plt.show()
```
