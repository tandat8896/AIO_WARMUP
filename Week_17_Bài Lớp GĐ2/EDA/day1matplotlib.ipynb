```python
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
```

```python
days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
temperatures = [22,24,25,23,26,28,27]
```

```python
fig,ax = plt.subplots()
ax.plot(days, temperatures,marker='o',color='r',linewidth = 2, linestyle='--')
ax.set_title("Weekly Temperature")
ax.set_xlabel("Day")
ax.set_ylabel("Temperature")
ax.grid(True)
plt.show()
```

```python
data = np.random.normal(0,1,1000)
plt.hist(data , bins=30,density=False,alpha = 0.7,color='g')
```

```python
data = np.random.normal(0,1,1000)
plt.hist(data , bins=30,density=True,alpha = 0.7,color='g') #density chuẩn hóa =1
```

```python
# Tạo dữ liệu ngẫu nhiên phân phối chuẩn
data = np.random.normal(0, 1, 1000)

# Vẽ biểu đồ histogram
counts, bin_edges, _ = plt.hist(data, bins=30, density=True, alpha=0.7, color='g')

# Vẽ các biên dưới (lower bin edges)
plt.vlines(bin_edges[:-1], 0, counts, colors='r', linestyles='dashed', label='Lower bin edges')
plt.vlines(bin_edges[1:], 0, counts, colors='b', linestyles='dashed', label='Upper bin edges')
# Hiển thị biểu đồ với biên dưới
plt.legend()
plt.show()
```

```python
x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,2,1)
ax.plot(x,y1,color='b')
ax.set_title('sin(x)')

ax2 = fig.add_subplot(1,2,2)
ax2.plot(x,y2,color='r')
ax.set_title('cos(x)')

plt.tight_layout()

plt.show()
```

```python
x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(1, 2, figsize=(10, 6))  # Tạo 1 hàng và 2 cột subplot
ax[0].plot(x, y1, color='r')  # Vẽ sin(x) lên subplot đầu tiên
ax[0].set_title('sin(x)')
```
