```python
import numpy as np

data = np.genfromtxt('/content/advertising.csv',
                    delimiter=',',skip_header=1)

print(data.shape)
data = data.tolist()
print(data)
TV=[]
radio=[]
newspaper=[]
sales=[]
for element in data:
  TV.append(element[0])
  radio.append(element[1])
  newspaper.append(element[2])
  sales.append(element[3])
print('TV',TV)
print('radio',radio)
print('newspaper',newspaper)
print('sales',sales)
```

```python
def predict(x1, x2, x3, w1, w2, w3, b):
    return w1*x1 + w2*x2 + w3*x3 + b

def compute_loss(y_hat, y):
    return (y_hat - y)**2

def compute_derivatives(x1, x2, x3, y_hat, y):
    dl_dw1 = 2*(y_hat - y)*x1
    dl_dw2 = 2*(y_hat - y)*x2
    dl_dw3 = 2*(y_hat - y)*x3
    dl_db  = 2*(y_hat - y)
    return dl_dw1, dl_dw2, dl_dw3, dl_db

def update(w1, w2, w3, b, lr, dl_dw1, dl_dw2, dl_dw3, dl_db):
    w1 = w1 - lr*dl_dw1
    w2 = w2 - lr*dl_dw2
    w3 = w3 - lr*dl_dw3
    b  = b  - lr*dl_db
    return w1, w2, w3, b
```

```python
w1 = 0.1
w2 = 0.1
w3 = -0.1
b = 0

# test case: sample 1 --> y_hat = 19.87
print(x1, x2, x3, y)
y_hat = predict(x1, x2, x3, w1, w2, w3, b)
print(y_hat)

# test: y = 22.1, y_hat = 19.87 --> loss = 4.9729
loss = compute_loss(y_hat, y)
print(loss)

# test: 230.1 37.8 69.2 ---> dl_dw1=-1026.246, dl_dw2, dl_dw3, dl_db=-4.46
dl_dw1, dl_dw2, dl_dw3, dl_db = compute_derivatives(x1, x2, x3, y_hat, y)
print(dl_dw1, dl_dw2, dl_dw3, dl_db)

# test: w1 = 0.1, lr=0.1, dl_dw1=-1026.246 ---> w1=?
w1, w2, w3, b = update(w1, w2, w3, b, 0.1, dl_dw1, dl_dw2, dl_dw3, dl_db)
print(w1, w2, w3, b)
```

```python
w1 = 0.1
w2 = 0.1
w3 = -0.1
b = 0
lr = 0.00001


losses = []
for _ in range(10):
    for i in range(200):
        # get data
        x1, x2, x3, y = TV[i], radio[i], newspaper[i], sales[i]

        # train
        y_hat = predict(x1, x2, x3, w1, w2, w3, b)
        loss = compute_loss(y_hat, y)
        losses.append(loss)

        dl_dw1, dl_dw2, dl_dw3, dl_db = compute_derivatives(x1, x2, x3, y_hat, y)
        w1, w2, w3, b = update(w1, w2, w3, b, lr, dl_dw1, dl_dw2, dl_dw3, dl_db)
```

```python
import matplotlib.pyplot as plt

plt.plot(losses)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.show()
```
